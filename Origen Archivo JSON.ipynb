Generar los 10,000 documentos y guardar en Google Drive

```python
from pymongo import MongoClient
from faker import Faker
import random
from datetime import datetime
from unidecode import unidecode  # Nueva librería para eliminar tildes

from google.colab import drive
drive.mount('/content/drive')

# Inicializar Faker para datos en español
fake = Faker('es_ES')

# Definir zonas y años
zonas = ['Sur', 'Norte', 'Centro', 'Oriente', 'Occidente']
años = [datetime.now().year - i for i in range(5)]

# Función para generar documentos radicados
def generar_documentos():
    documentos = []
    tipos_documentos = ['Contrato', 'Factura', 'Recibo', 'Identificacion', 'Poder Notarial']  # Sin tildes
    num_docs = random.randint(1, 5)
    for _ in range(num_docs):
        documentos.append(random.choice(tipos_documentos))
    return documentos

# Función para generar número de celular colombiano
def generar_numero_celular_colombiano():
    # Estructura típica: 10 dígitos, iniciando con '3' (por ejemplo, 300, 310, 320, etc.)
    prefijos = ['300', '310', '320', '350', '315', '312', '314', '317', '321']  # Algunos prefijos válidos
    prefijo = random.choice(prefijos)
    numero = ''.join([str(random.randint(0, 9)) for _ in range(7)])  # Genera los 7 dígitos restantes
    return f"+57 {prefijo}{numero}"

# Generar 10,000 documentos
documentos_a_insertar = []
for _ in range(10000):
    año = random.choice(años)
    mes = random.randint(1, 12)
    dia = random.randint(1, 28)  # Para evitar errores con febrero
    fecha_separacion = datetime(año, mes, dia)

    documento = {
        'nombre_cliente': unidecode(fake.name()),  # Elimina tildes
        'numero_celular': generar_numero_celular_colombiano(),
        'fecha_separacion': fecha_separacion.strftime('%Y-%m-%d'),
        'valor_separacion': round(random.uniform(5000, 50000), 2),
        'valor_inmueble': round(random.uniform(50000, 500000), 2),
        'valor_credito': round(random.uniform(10000, 300000), 2),
        'documentos_radicados': ', '.join(generar_documentos()),  # Convertir a una sola cadena de texto
        'zona': random.choice(zonas),
        'año': año
    }
    documentos_a_insertar.append(documento)

# Guardar los documentos en un archivo JSON sin tildes
import json

file_path = '/content/drive/My Drive/documentos_constructorabolivar_sin_tildes.json'
with open(file_path, 'w', encoding='utf-8') as f:
    json.dump(documentos_a_insertar, f, ensure_ascii=False, indent=4)

print(f'Archivo guardado en {file_path}')
```
